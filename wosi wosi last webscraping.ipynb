```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import urllib.parse
```

```python
website = 'https://wosiwosi.co.uk/product-category/supermarket/snacks-and-confectionaries/'
```

```python
response = requests.get(website)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

```python
results = soup.find_all('div', {'class':'astra-shop-summary-wrap'})
```

```python
len(results)
```

```python
results[0]
```

```python
results[0].find('h2').get_text()
```

```python
results[0].find('span', {'class': 'price'}).get_text()
```

```python
name = []
price = []

for result in results:
    
    #name
    try:
        name.append(result.find('h2').get_text())
    except:
        name.append('n/a')
        
    #price
    try:
        price.append(result.find('span', {'class': 'price'}).get_text())
    except:
        price.append('n/a') 
```

```python
price_overview = pd.DataFrame({'Name': name, 'Price': price })
```

```python
price_overview
```

```python
price_overview.to_excel('results_confectionaries_page.xlsx', index=False)
```
