```python
name = []
price = []

for result in results:
    
    #name
    try:
        name.append(result.find('h2').get_text())
    except:
        name.append('n/a')
        
    #price
    try:
        price.append(result.find('span', {'class': 'price'}).get_text())
    except:
        price.append('n/a')        
```

```python
price_overview = pd.DataFrame({'Name': name, 'Price': price })
```

```python
name = []
price = []

for i in range(1,4):
    
    #website in a variable
    website = 'https://wosiwosi.co.uk/product-category/grain/page/' + str(i)
    
    #request
    response = requests.get(website)
    
    # soup object
    soup = BeautifulSoup(response.content,'html.parser')
        
    #results
    results = soup.find_all('div', {'class':'astra-shop-summary-wrap'})
    
    for result in results:

        #name
        try:
            name.append(result.find('h2').get_text())
        except:
            name.append('n/a')

        #price
        try:
            price.append(result.find('span', {'class': 'price'}).get_text())
        except:
            price.append('n/a')        
```

```python
price_overview = pd.DataFrame({'Name': name, 'Price': price })
```
