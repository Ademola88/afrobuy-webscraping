```python
website = 'https://johnandbiola.co.uk/world-food'
```

```python
response = requests.get(website)
```

```python
name = []
price = []

for result in results:
    
    #name
    try:
        name.append(result.find('h2', {'class': 'product-title'}).get_text().strip('\n'))
    except:
        name.append('n/a')
        
     #price
    try:
        price.append(result.find('span', {'class': 'price only-price'}).get_text().strip('From'))
    except:
        price.append('n/a')
```

```python
price_overview = pd.DataFrame({'Name': name, 'Price': price })
```

```python
name = []
price = []

for i in range(1,5):
    
    #website in a variable
    website = 'https://johnandbiola.co.uk/world-food?pagenumber=' + str(i) + '&viewmode=&pagesize=20&orderby=0'
    
    #request
    response = requests.get(website)
    
    # soup object
    soup = BeautifulSoup(response.content,'html.parser')
        
    #results
    results = soup.find_all('div', {'class':'details'})
    
    for result in results:
    
        #name
        try:
            name.append(result.find('h2', {'class': 'product-title'}).get_text().strip('\n'))
        except:
            name.append('n/a')

         #price
        try:
            price.append(result.find('span', {'class': 'price only-price'}).get_text().strip('From'))
        except:
            price.append('n/a')
```

```python
price_overview = pd.DataFrame({'Name': name, 'Price': price })
```

```python
price_overview
```

```python
price_overview.to_excel('results_johnbiolaworld_page.xlsx', index=False)
```
