```python
name = []
price = []

for result in results:
    
    #name
    try:
        name.append(result.find('h5').get_text())
    except:
        name.append('n/a')
        
    #price
    try:
        price.append(result.find('div', {'class': 'price'}).get_text())
    except:
        price.append('n/a') 
```

```python
price_overview = pd.DataFrame({'Name': name, 'Price': price })
```

```python
name = []
price = []

for i in range(1,3):
    
    #website in a variable
    website = 'https://afrobuy.co.uk/collections/condiments?page=' + str(i)
    
    #request
    response = requests.get(website)
    
    # soup object
    soup = BeautifulSoup(response.content,'html.parser')
        
    #results
    results = soup.find_all('div', {'class':'desc'})
    
    for result in results:

        #name
        try:
            name.append(result.find('h5').get_text())
        except:
            name.append('n/a')

        #price
        try:
            price.append(result.find('div', {'class': 'price'}).get_text())
        except:
            price.append('n/a')        
```

```python
price_overview = pd.DataFrame({'Name': name, 'Price': price })
```

```python
price_overview
```
