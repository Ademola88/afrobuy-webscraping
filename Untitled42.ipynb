```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import urllib.parse
```

```python
website = 'https://johnandbiola.co.uk/oils'
```

```python
response = requests.get(website)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

```python
results = soup.find_all('div', {'class':'details'})
```

```python
len(results)
```

```python
results[0]
```

```python
results[0].find('h2', {'class': 'product-title'}).get_text().strip('\n')
```

```python
results[0].find('span', {'class': 'price only-price'}).get_text().strip('From')
```

```python
name = []
price = []

for result in results:
    
    #name
    try:
        name.append(result.find('h2', {'class': 'product-title'}).get_text().strip('\n'))
    except:
        name.append('n/a')
        
     #price
    try:
        price.append(result.find('span', {'class': 'price only-price'}).get_text().strip('From'))
    except:
        price.append('n/a')
```

```python
price_overview = pd.DataFrame({'Name': name, 'Price': price })
```

```python
price_overview
```

```python
price_overview.to_excel('results_johnbiolaoil_page.xlsx', index=False)
```
