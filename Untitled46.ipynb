```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
website = 'https://spiceeupp.co.uk/18-snacks?id_category=18&n=69'
```

```python
response = requests.get(website)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

```python
results = soup.find_all('div', {'class':'right-block'})
```

```python
len(results)
```

```python
results[0]
```

```python
results[0].find('h5', {'class': 'product-name'}).get_text()
```

```python
results[0].find('span', {'class': 'price product-price'}).get_text()
```

```python
name = []
price = []

for result in results:
    
    #name
    try:
        name.append(result.find('h5', {'class': 'product-name'}).get_text())
    except:
        name.append('n/a')
        
     #price
    try:
        price.append(result.find('span', {'class': 'price product-price'}).get_text())
    except:
        price.append('n/a')
```

```python
price_overview = pd.DataFrame({'Name': name, 'Price': price })
```

```python
price_overview
```

```python
price_overview.to_excel('results_spiceeuppsnack_page.xlsx', index=False)
```
