```python
name = []
price = []

for i in range(1,4):
    
    #website in a variable
    website = 'https://www.royacshop.com/shop/health-beauty/page/' + str(i)
    
    #request
    response = requests.get(website)
    
    # soup object
    soup = BeautifulSoup(response.content,'html.parser')
        
    #results
    results = soup.find_all('div', {'class':'box-text box-text-products'})
    
    for result in results:
    
        #name
        try:
            name.append(result.find('div', {'class': 'title-wrapper'}).get_text().strip('Food Items'))
        except:
            name.append('n/a')

         #price
        try:
            price.append(result.find('span', {'class': 'price'}).get_text())
        except:
            price.append('n/a')
```

```python
price_overview = pd.DataFrame({'Name': name, 'Price': price })
```

```python
price_overview
```
